## 1. Conditional Statements (`if`, `elif`, `else`)
Python executes code blocks based on whether a condition is `True` or `False`.

* **Syntax:**
    ```python
    if condition_1:
        # Runs if condition_1 is True
    elif condition_2:
        # Runs if condition_1 was False AND condition_2 is True
    else:
        # Runs if ALL above were False
    ```

## 2. Boolean Logic
Logic gates return a `bool` type: `True` or `False`.

### Comparison Operators
| Symbol | Meaning | Example | Result |
| :--- | :--- | :--- | :--- |
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal | `5 != 3` | `True` |
| `>` | Greater than | `5 > 10` | `False` |
| `>=` | Greater/Equal | `5 >= 5` | `True` |

### Logical Operators
| Symbol | Meaning | Example |
| :--- | :--- | :--- |
| `and` | Both must be True | `(x > 0) and (x < 10)` |
| `or` | At least one is True | `(x == 0) or (x == 1)` |
| `not` | Reverses the result | `not (5 > 3)` ‚Üí `False` |

> ‚ö†Ô∏è **The Assignment Trap:**
> * `a = 5` Sets `a` to 5 (**Assignment**)
> * `a == 5` Asks "Is `a` 5?" (**Comparison**)

## 3. The `while` Loop (Conditional)
Repeats a block of code as long as a condition remains `True`.

* **Use Case:** When you don't know how many times you need to loop (e.g., finding a root, waiting for user input).
* **Risk:** Infinite loops. Ensure the condition eventually becomes `False`.

```python
x = 0
while x < 5:
    print(x)
    x += 1  # Crucial: Update x so the loop eventually ends
